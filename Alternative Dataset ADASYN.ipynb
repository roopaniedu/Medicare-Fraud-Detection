# Data Processing

```python
train_df = pd.read_csv("/content/drive/MyDrive/BT4012 Project/Submission Folder/train_df.csv")
test_df = pd.read_csv("/content/drive/MyDrive/BT4012 Project/Submission Folder/test_df.csv")
```

```python
# highly imbalanced dataset
print(sorted(Counter(train_df['Is_Fraud']).items()))
```

```python
X_train, y_train = train_df.drop(columns=["NPI", "Is_Fraud"]), train_df["Is_Fraud"]
X_test, y_test = test_df.drop(columns=["NPI", "Is_Fraud"]), test_df["Is_Fraud"]

print(sorted(Counter(y_train).items())) # training set is also very imbalanced
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = pd.DataFrame(scaler.fit_transform(X_train), columns=X_train.columns)
X_test = pd.DataFrame(scaler.transform(X_test), columns=X_test.columns)
```

```python
X_train_svm_smote = pd.read_csv('/content/drive/MyDrive/BT4012 Project/Submission Folder/X_train_svm_smote.csv')
X_train_svm_smote_under = pd.read_csv('/content/drive/MyDrive/BT4012 Project/Submission Folder/X_train_svm_smote_under.csv')
X_train_smote_enn = pd.read_csv('/content/drive/MyDrive/BT4012 Project/Submission Folder/X_train_smote_enn.csv')
X_train_adasyn = pd.read_csv('/content/drive/MyDrive/BT4012 Project/Submission Folder/X_train_adasyn.csv')

y_train_svm_smote_under = pd.read_csv('/content/drive/MyDrive/BT4012 Project/Submission Folder/y_train_svm_smote_under.csv')
y_train_svm_smote = pd.read_csv('/content/drive/MyDrive/BT4012 Project/Submission Folder/y_train_svm_smote.csv')
y_train_smote_enn = pd.read_csv('/content/drive/MyDrive/BT4012 Project/Submission Folder/y_train_smote_enn.csv')
y_train_adasyn = pd.read_csv('/content/drive/MyDrive/BT4012 Project/Submission Folder/y_train_adasyn.csv')
```

```python
adasyn_over_sample = ADASYN(sampling_strategy='minority', random_state=8, n_neighbors=3)
X_train_adasyn, y_train_adasyn = adasyn_over_sample.fit_resample(X_train, y_train)
print(sorted(Counter(y_train_adasyn).items())) # training set is also very imbalanced
```

```python
X_train_adasyn.to_csv("/content/drive/MyDrive/BT4012 Project/Submission Folder/X_train_adasyn.csv", index=False)
y_train_adasyn.to_csv("/content/drive/MyDrive/BT4012 Project/Submission Folder/y_train_adasyn.csv", index=False)
```
